```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import cx_Oracle
from fbprophet import Prophet
from fbprophet.diagnostics import cross_validation
from fbprophet.diagnostics import performance_metrics
```

```python
dsn_tns = cx_Oracle.makedsn('racdclscan.cardeal.corp', '1521', service_name='winthor') # if needed, place an 'r' before any parameter in order to address special characters such as '\'.
conn = cx_Oracle.connect(user=r'CONSULTA', password='CONSULTA', dsn=dsn_tns) # if needed, place an 'r' before any parameter in order to address special characters such as '\'. For example, if your user name contains '\', you'll need to place 'r' before the user name: user=r'User Name'
```

```python
df = pd.read_sql_query(
'''
SELECT
DATA,
sum(vlitemsemst) vd

FROM WINTHOR.xv_ped
group by data

''', conn)

holidays = pd.read_excel('holidays.xlsx')
```

```python
df.columns = ['ds', 'y']

#removing weekends
df = df[df['ds'].dt.dayofweek < 5]
df = df[~(df['ds'].isin(list(holidays['ds'])))]

df.dropna(inplace = True)
df.head()


prediction_size = 100

# Initialize and train a model
m = Prophet(growth = 'linear',
    daily_seasonality = False,
    weekly_seasonality = False,
    yearly_seasonality = False).add_seasonality(
    name = 'weekly', period = 7, fourier_order = 20).add_seasonality(
    name = 'monthly', period = 30.5, fourier_order = 20).add_seasonality(
    name = 'yearly', period = 365.25, fourier_order = 17) 

m.fit(df)

# Make predictions
future = m.make_future_dataframe(periods=prediction_size)
future = future[future['ds'].dt.dayofweek < 5]
future = future[~(future['ds'].isin(list(holidays['ds'])))]

forecast = m.predict(future)
forecast.head()

# Plot forecast
m.plot(forecast)

# Plot forecast's components
m.plot_components(forecast)

# Evaluate the model
def make_comparison_dataframe(historical, forecast):
    return forecast.set_index('ds')[['yhat', 'yhat_lower', 'yhat_upper']].join(historical.set_index('ds'))

cmp_df = make_comparison_dataframe(df, forecast)

cmp_df.to_excel('cmp.xlsx')
```

```python
cv_results = cross_validation(m, horizon = '30 days') 

cv_results = cv_results[(cv_results['y'] != 0)]

def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape_baseline = mean_absolute_percentage_error(cv_results.y, cv_results.yhat)
mape_baseline
```
